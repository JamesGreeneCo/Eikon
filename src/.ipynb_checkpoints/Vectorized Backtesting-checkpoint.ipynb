```python
ric = 'AAPL.O'
```

```python
data = ek.get_timeseries(ric, fields = 'CLOSE', start_date = '2010-01-01', end_date = '2018-04-30')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.normalize().iplot(kind='lines')
```

```python
SMA1 = 42
SMA2 = 252
```

```python
data['SMA1'] = data['CLOSE'].rolling(SMA1).mean()
data['SMA2'] = data['CLOSE'].rolling(SMA2).mean()
```

```python
data.head()
```

```python
data.tail()
```

```python
data.iplot()
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.iplot()
```

```python
data['POSITIONS'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
```

```python
data.iplot(secondary_y = 'POSITIONS')
```

```python
data['RETURNS'] = np.log(data['CLOSE'] / data['CLOSE'].shift(1))
```

```python
data['RETURNS'].head()
```

```python
data.dropna(inplace=True)
```

```python
data['RETURNS'].iplot(kind='histogram', subplots=True)
```

```python
data['STRATEGY'] = data['POSITIONS'].shift(1) * data['RETURNS']
```

```python
data['STRATEGY'].head()
```

```python
data.dropna(inplace=True)
```

```python
np.exp(data[['RETURNS', 'STRATEGY']].sum())
```

```python
data[['RETURNS', 'STRATEGY']].cumsum().apply(np.exp).iplot()
```
