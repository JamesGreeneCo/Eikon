```python
import eikon as ek
import cufflinks as cf
import configparser as cp
import numpy as np
cf.set_config_file(offline=True)
```

```python
with open('../key.txt', 'r') as file:
    key = file.read()
print(key)
```

```python
ek.set_app_key(key)
```

```python
apple = ek.get_timeseries('AAPL.O', fields = ['CLOSE', 'HIGH', 'LOW', 'OPEN'], start_date = '2019-01-03 09:30:00', end_date = '2019-01-3 16:00:00', interval = 'minute')
```

```python
apple.info()
```

```python
apple.iplot(kind='candle')
```

```python
SMA1 = 80
SMA2 = 250
```

```python
apple['SMA1'] = apple['CLOSE'].rolling(SMA1).mean()
apple['SMA2'] = apple['CLOSE'].rolling(SMA2).mean()
apple.head()
```

```python
apple.dropna(inplace=True)
```

```python
apple
```

```python
apple.normalize().iplot(kind='candle', secondary_y='SMA1')
```
